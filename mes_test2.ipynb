```python
print("SUMMARY:", data.get("summary"))
print("CREATED:", data.get("created"))

assignee = data.get("assignee")
if assignee:
    print("ASSIGNEE:", assignee.get("fullName"))
else:
    print("ASSIGNEE: None")
```

```python
print("\nCUSTOM FIELDS:\n")

for field in data.get("customFields", []):
    name = field.get("name")
    value = field.get("value")

    if isinstance(value, dict):
        # Handle structured values
        if "name" in value:
            print(f"{name}: {value.get('name')}")
        elif "minutes" in value:
            print(f"{name}: {value.get('minutes')} minutes")
        else:
            print(f"{name}: {value}")
    else:
        print(f"{name}: {value}")
```

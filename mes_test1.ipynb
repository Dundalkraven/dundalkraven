```python
import requests
from docxtpl import DocxTemplate
from flask import Flask


# =========================
# CONFIG
# =========================
ISSUE_ID = "sysdes-30009"
YOUTRACK_URL = f"https://youtrack.meteocontrol.de/api/issues/{ISSUE_ID}"
TOKEN = "perm-ci5oYW1pbGk=.NzctMTM3.2DmnYBuMYzQf04AlOrYtisyZago3Rm"

TEMPLATE_FILE = "template.docx"
OUTPUT_FILE = "test_report.docx"

# =========================
# FETCH DATA FROM YOUTRACK
# =========================
headers = {
    "Authorization": f"Bearer {TOKEN}",
    "Accept": "application/json"
}

params = {
    "fields": "summary,customFields(name,value(name)),assignee(fullName)"
}

response = requests.get(YOUTRACK_URL, headers=headers, params=params)

if response.status_code != 200:
    print(" API Error:", response.text)
    exit()

issue = response.json()

# =========================
# EXTRACT FIELDS
# =========================
def extract_data(issue):
    project_name = issue.get("summary")
    assignee = issue.get("Assignee", {}).get("fullName")

    project_number = None

    for field in issue.get("customFields", []):
        if field.get("name") == "Project Number":
            value = field.get("value")

            if isinstance(value, dict):
                project_number = value.get("name")

            elif isinstance(value, str):
                project_number = value

            elif isinstance(value, int):
                project_number = str(value)

    return {
        "project_name": project_name,
        "Intervenant": MISE EN SERVICE REALISEE PAR,
        "project_number": project_number,
        "TOTAL 0H" : spent_time,
        "DUREE OPERATIONS A REALISER ESTIMEE" : Duree
    }

data = extract_data(issue)

print("Extracted:", data)

# =========================
# GENERATE WORD REPORT
# =========================
doc = DocxTemplate(TEMPLATE_FILE)

# The keys on the LEFT must be IDENTICAL to what is inside {{ }} in Word
context = {
    "Project_Name": data["project_name"],
    "Project_Number": data["project_number"],
    "Intervenant": data["assignee"], # Maps assignee to your Intervenant tag
    "TOTAL 0H": "0h" # Added to prevent errors if the tag exists
}

print("Rendering with context:", context)

doc.render(context)
doc.save(OUTPUT_FILE)

print("✅ Report generated:", OUTPUT_FILE)
```
